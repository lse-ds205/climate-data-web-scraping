```python
import scrapy
from scrapy.http import Request
from scrapy.selector import Selector
import requests
url = "https://climateactiontracker.org/countries/india/"
response = requests.get(url)
selector = Selector(text=response.text)
```

```python
data = []
```

country Name

```python
data.append(selector.css('h1::text').get())
selector.css('h1::text').get()
```

Overall rating

```python
data.append(selector.css('dd::text').get())
selector.css('dd::text').get()
```

### Indicators Row 2

each indicator stored in its own tile

```python
tiles_row2 = selector.css('.ratings-matrix__second-row-cell')
```

we can iterate through each tile to extract data points

```python

indicator_data = []
for tile in tiles_row2:

    term = tile.css('dl dt::text').get()

    if not term:
        ## most indicators stored under p tags under dt tag but not all
        term= tile.css('dl dt p::text').get()

    term_details = tile.css("i ::text").get()


    value = tile.css('dl dd b::text').get()

    metric = tile.css("dd i ::text").get()
    tile_contents =  {"term": term,"term_details": term_details, "value": value,"metric":metric}

    
    indicator_data.append(tile_contents)



# Print extracted data
for item in indicator_data:
    print(f"{item['term']}, {item["term_details"]}: {item['value']}, {item["metric"]}")
```

### Indicators row 3 Net zero target 

```python
net_zero_tile = selector.css('.ratings-matrix__third-row')
```

```python

term = net_zero_tile.css('dl dt::text').get()




bold_values = net_zero_tile.css('dl dd b::text').getall()
year = bold_values[0].strip()

value = bold_values[1].strip()

metric = net_zero_tile.css('dl dd p::text').getall()[1]

tile_contents =  {"term": term,"term_details": year, "value": value,"metric":metric}

indicator_data.append(tile_contents)
```

### (unnofficial) fourth row tile land use and forestry

```python
term = selector.css('.ratings-matrix__land_use_forestry dt::text').get().strip()
value = selector.css('.ratings-matrix__land_use_forestry dd b::text').get().strip()

tile_contents =  {"term": term,"term_details": None, "value": value,"metric":None}
indicator_data.append(tile_contents)
```

```python
data.append(indicator_data)
data
```
